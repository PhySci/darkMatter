```python
# predict position of the track on the next layer
def getPosition(track,dz):
    '''
    track - track record (X,Y,Z,TX,TY)
    '''
    pos = np.array([track.X, track.Y])
    pos[0] +=  dz*np.tan(track.Tx)
    pos[1] +=  dz*np.tan(track.TY)  
    return pos
```
